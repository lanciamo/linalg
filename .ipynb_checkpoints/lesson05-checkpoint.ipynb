__2.__ Дан оператор поворота на 180 градусов, задаваемый матрицей 

$$A=\begin{pmatrix}
-1 & 0\\ 
0 & -1
\end{pmatrix}.$$

Показать, что __любой__ вектор является для него собственным.

```python
''''
в этом случае должно выполняться равенство в двух уравнениях
-1*x + 0*y = z*x; z = -1
0*x + -1*y = z*y; z = -1
соответсвенно, при любых значениях x, y равенство выполняется, соответсвенно любой вектор - собственный
```

__3.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
1 & 1\\ 
-1 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(1,1)$ собственным вектором этого линейного оператора.

```python
''''
в этом случае должно выполняться равенство в двух уравнениях
1*1 + 1*1 = x; x = 2
-1*1 + 3*1 = x; x = 2
соответсвенно, да - этот вектор является собственным
```

__4.__ Пусть линейный оператор задан матрицей

$$A=\begin{pmatrix}
0 & 3 & 0\\ 
3 & 0 & 0\\
0 & 0 & 3
\end{pmatrix}.$$

Установить, является ли вектор $x=(3, -3, -4)$ собственным вектором этого линейного оператора.

```python
''''
в этом случае должно выполняться равенство в трех уравнениях
0*3 + 3*-3 + 0*-4 = 3x; x = -3
3*3 + -3*0 + 0*-4 = -3x; x = -3
0*3 + -3*0 + 3*-4 = -4x; x = 3
соответсвенно, нет - этот вектор не является собственным

```
